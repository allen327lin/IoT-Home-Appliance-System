```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore

# Use a service account
cred = credentials.Certificate('/content/drive/MyDrive/NTUT/專題 - IoT/Firebase金鑰/ntut-iot-d5b34-firebase-adminsdk-sj4yb-0b2d0ef3a8.json')
firebase_admin.initialize_app(cred)

db = firestore.client()
```

# Write

```python
doc_ref = db.collection(u'All').document(u'doc1')
doc_ref.set({
    u'name': u'Allen',
    u'born': 2000
})
```

```python
doc_ref = db.collection(u'abc').document(u'123')
doc_ref.set({
    u'school': u'NTUT',
    u'?': u'IAE'
})
```

# Read

```python
users_ref = db.collection(u'All')
docs = users_ref.stream()

for doc in docs:
    print(f'{doc.id} => {doc.to_dict()}')
```
